```python
from paddleocr import PaddleOCR, draw_ocr
import matplotlib.pyplot as plt
from PIL import Image

# Initialize PaddleOCR
ocr = PaddleOCR(use_angle_cls=True, lang='en')

# Specify your image path
img_path = '/Users/402-07/YoungMan_projectDE/receipt_list/32.jpg'

# Perform OCR
result = ocr.ocr(img_path, cls=True)

# Load the image using PIL
image = Image.open(img_path).convert('RGB')

# Draw the OCR results on the image
boxes = [line[0] for line in result]  # Extracting bounding boxes
txts = [line[1][0] for line in result]  # Extracting detected texts
scores = [line[1][1][0] if isinstance(line[1][1], tuple) else line[1][1] for line in result]
im_show = draw_ocr(image, boxes, txts, scores)

# Convert to a format suitable for matplotlib
im_show = Image.fromarray(im_show)

# Display the image
plt.figure(figsize=(10,10))
plt.imshow(im_show)
plt.axis('off')
plt.show()

# Print the OCR results
for line in result:
    print(line)
```
