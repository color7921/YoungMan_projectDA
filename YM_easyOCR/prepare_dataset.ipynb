```python
from PIL import Image
import json

from torchvision import transforms, datasets
from torch.utils.data import DataLoader

name = "IMG_OCR_53_4PR_09181"
# JSON 파일을 불러와서 파싱
with open(f'C:/Users/402-07/Downloads/handwritten/data/Training_labeling/{name}.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# 원본 이미지 불러오기
image_path = f'C:/Users/402-07/Downloads/handwritten/data/T_image/Training_image/{name}.png'  # 이미지 파일 이름을 정확히 입력해야 합니다.
image = Image.open(image_path)
i = 0
list = []
# 이미지에서 각 박스를 잘라내고 라벨과 매핑
for box in data['bbox']:
    i = i + 1
    # 박스 좌표 추출
    x_coordinates = box['x']
    y_coordinates = box['y']
    left, top, right, bottom = min(x_coordinates), min(y_coordinates), max(x_coordinates), max(y_coordinates)

    # 이미지 자르기
    cropped_image = image.crop((left, top, right, bottom))

    # 잘라낸 이미지와 라벨 매핑 (예: cropped_image를 저장하고, 해당 라벨을 출력 또는 저장)
    print(box['data'])  # 라벨 출력
    label = box['data']
    cropped_image.save(f'C:/Users/402-07/Downloads/handwritten/data/cropped/{i}{name}.png')  # 잘라낸 이미지 저장

    # transform = transforms.Compose([
    #     transforms.Resize((37, 256)),  # Resize to 256x256 pixels
    #     transforms.ToTensor()
    # ])

    # dataset = datasets.ImageFolder(root='path/to/your/dataset', transform=transform)
    
    fileDir = f"cropped/{i}{name}.png"
    list.append(f'{fileDir}\t{label}\n')
with open('C:/Users/402-07/Downloads/handwritten/data/trained/gt.txt', 'w', encoding='utf-8') as file:
    for data in list:
        file.write(data)
    

# 필요에 따라 잘라낸 이미지와 라벨을 사용하여 추가 작업 수행
```
